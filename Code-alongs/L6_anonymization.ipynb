```python
import hashlib as hl


secret = "gore_bord@gmail.com".encode()  # utf-8 encoded version
secret2 = "gore_bord1@gmail.com".encode()

print(secret)

secret_hash = hl.sha256(secret).hexdigest()
secret_hash2 = hl.sha256(secret2).hexdigest()

print(f"secret_hash : {secret_hash}")

secret_hash_same= hl.sha256("gore_bord@gmail.com".encode()).hexdigest()

print(f"secret_hash_same : {secret_hash_same}")
print(f"secret_hash2 : {secret_hash2}")

secret_hash == secret_hash_same
```

# Exemple with Pandas


```python
import pandas as pd 

bank = pd.read_csv("Data/BankChurners.csv")

bank.head(3)
```

```python
bank["CLIENTNUM"] = bank["CLIENTNUM"].astype(str)

bank["CLIENTNUM"].head()
```

```python
hash_series = bank["CLIENTNUM"].apply(lambda x: hl.sha256(x.encode()).hexdigest())

hash_series.head()
```

```python
bank.insert(1,"Hashed client numbers",hash_series)

bank.head()
```

```python
bank = bank.drop(columns="CLIENTNUM")

bank.head()
```
